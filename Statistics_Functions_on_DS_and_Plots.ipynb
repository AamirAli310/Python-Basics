Goind to remove null value base records to have better annalysis


```python
nonnullSDS = salesDS.dropna(subset=['market_value_mil','profit_mil'])
nonnullSDS.info()
```

```python

nonnullSDS.isnull().sum()
```

```python
nonnullSDS.describe()
```

Finding mean on revenue_mil

```python
nonnullSDS['revenue_mil'].mean()
```

Group dataset on which we want to perform observation

```python
mySDS = nonnullSDS[['rank','year','revenue_mil','asset_mil']]
print("Mean:",mySDS['revenue_mil'].mean())
print("Median:",mySDS['revenue_mil'].median())
print("Std:",mySDS['revenue_mil'].std())
print("Variance:",mySDS['revenue_mil'].var())
print("Min:",mySDS['revenue_mil'].min())
print("Max:",mySDS['revenue_mil'].max())
print("Count:",mySDS['revenue_mil'].count())
Q1 = mySDS['revenue_mil'].quantile(0.25);
print("Quartile Q1:", Q1)
Q2 = mySDS['revenue_mil'].quantile(0.50)
print("Quartile Q2:",Q2)
Q3 = mySDS['revenue_mil'].quantile(0.75)
print("Quartile Q3:",Q3)
print("Finding Inter Quantile Range");
print("IQR: Q3 - Q1")
print("IQR=",Q3-Q1)
```

Drawing Bell shape for the Data analysis to verify either it's normilized or Left or Right SKEWED

```python
#mySDS['revenue_mil'].plot(kind='kde', ylabel="Volume")

#Ploting using seaborn library
import seaborn as sns
sns.distplot(mySDS['revenue_mil'], bins=15, hist_kws={'alpha': 0.4, 'color':'green'}, color="red")
#sns.histplot(mySDS['revenue_mil'], color="orange", binwidth=4)
```
