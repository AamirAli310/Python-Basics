This is my Util class to be use in next modules

```python
class Utils:
  def __init__(self):
    print("Util constructor fired");

#************ MY UTILITY FUNCTIONS START ***************

#function to take user inputs as per user message & data type
  def getUserInput(message, type):
    userInput = type(input(message));
    return userInput;

#Function to print all keys of a dictionary dynamically
  def printDictionary(dic):
    #For loop on Dict's keys array
    for key in dic.keys():
      if len(key) > 4:
        print("Item- ", key, ":\t",dic[key]);
      else:
        print("Item- ", key, ":\t\t",dic[key])
      print("\n-----------------------------");

#************ MY UTILITY FUNCTIONS END ***************
  def __main__(self):
    print("Util main loaded");
```

Idea is to show list of Plotlib common functions as dropdown/options beside imported excel/csv data source attributes which user can select and see the graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

myDS = pd.read_csv("/content/AA_Data/adultdata.csv")
charts = ["Bar","Histogram","Pie","Plot","Scatter"]
usrAction = Utils.getUserInput("Do you want to continue the program? Y:Yes & N: No ", str).lower()
uColX = uColY = selectedChart = ""

while usrAction == "y":
  dsCols = np.array(myDS.columns);
  print("Choose a chart from the given list: \n",charts)
  while selectedChart.title() not in charts:
    selectedChart = Utils.getUserInput("::> ", str).lower()
  plt.grid()
  chartFunRef = getattr(plt, selectedChart)
  if selectedChart != 'plot': #Using while loop for condition to re-take input from the user until a postive number is entered
    print("Here is Header of imported data : \n", dsCols)
    while uColX not in dsCols:
      uColX = Utils.getUserInput("Choose column name for X-Axis. ", str).lower()

    print("Here is Header of imported data : \n", dsCols)
    while uColY not in dsCols:
      uColY = Utils.getUserInput("Choose column name for Y-Axis. ", str).lower()

    chartFunRef(myDS[uColX], myDS[uColY], color="green")
  else:
    while uColX not in dsCols:
      uColX = Utils.getUserInput("Choose column name. ", str).lower()
    chartFunRef(myDS[uColX], marker="*", color="red")

  plt.show()
  #Resetting the parameters
  selectedChart = uColX = uColY = ""
  usrAction = Utils.getUserInput("Do you want to continue the program?Y:Yes & N: No ", str).lower()
print("\n\n Thanks for using AA Utils Tool")
```
