```python
#importing more necessory libraries
import torch #For Deep learning library to work with yolov5
import cv2 # for image processing
from google.colab.patches import cv2_imshow # for image demonstration
```

```python
#Loading now pretrained model with the help of Torch lib to inject Yolov5 from Yolo lib
model  = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
imgPath = '/content/bicyleman.png'
image = cv2.imread(imgPath) #Oeiginal image
cv2_imshow(image)
```

```python
#Working on image with model
results = model(imgPath) #Here we don't have to call model.fit, for yolo we need to pass image path into model straight
```

```python
#To get image from output of the model results
images_box = results.render()[0] #output image from 1st index
cv2_imshow(images_box)#Now it will show the box plots with labelling to images after object detections
```
